```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# pull in airbnb data csv
file = "airbnb_q419_to_q420.csv"

airbnb_data = pd.read_csv(file)
airbnb_data
```

```python
#citation: https://stackoverflow.com/questions/24082784/pandas-dataframe-groupby-datetime-month
airbnb_data.index = pd.to_datetime(airbnb_data['date'],format="%m/%d/%y")
airbnb_data["month"] = airbnb_data.index.month
airbnb_data["year"] = airbnb_data.index.year
airbnb_data["final_date"] = airbnb_data["month"].astype(str)+"/" + airbnb_data["year"].astype(str)
airbnb_data
```

```python
airbnb_grouped = airbnb_data.groupby("final_date")
airbnb_grouped.count()
plt.figure(figsize = (15,10))
plt.plot(airbnb_grouped["listing_id"].count())
plt.show()
```
